# Two Interfaces for the Price of One

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

### Object-oriented interface


```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig,ax=plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig,ax=plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x,np.sin(x-2),'k')
ax[0].plot(x,np.sin(x),'g')
plt.plot(x,np.sin(x-1),'r')
ax[1].plot(x,np.cos(x),'y')
```

# Adjusting the Plot: Line Colors and Styles


```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:

plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
# rgbcmykw color standard abbrevetions
```

# Matplotlib Gotchas

While most plt functions translate directly to ax methods (such as plt.plot()  ax.plot(), plt.legend()  ax.legend(), etc.), this is not the case for all commands. In particular, functions to set limits, labels, and titles are slightly modified.

For transitioning between MATLAB-style functions and object oriented methods,make the following changes:
 plt.xlabel()  ax.set_xlabel()

 plt.ylabel()  ax.set_ylabel()

 plt.xlim()  ax.set_xlim()

 plt.ylim()  ax.set_ylim()

 plt.title()  ax.set_title()

In the object-oriented interface to plotting, rather than calling these functions individually, it is often more convenient to use the ax.set() method to set all these properties at once 

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),xlabel='x', ylabel='sin(x)',title='A Simple Plot');
```

# Bar Plot

```python
# plt.bar(x,height,width,color,label,bottom,align,edgecolor,alpha,linestyle,linewidth,visible)
```

```python
classes=['python','R','Java','C','C++']
classes_a=np.array([34,45,25,46,57])
classes_b=np.array([23,46,74,38,95])
classes_c=np.array([27,74,46,63,28])
```

```python
plt.bar(classes,classes_a)
```

```python
plt.barh(classes,classes_a)
```

```python
plt.bar(classes,classes_a)
plt.bar(classes,classes_b)
```

```python
n=np.arange(len(classes))
width=0.2
plt.bar(n,classes_a,width=width)
plt.bar(n+width,classes_b,width=width)
plt.bar(n+width+width,classes_c,width=width)
plt.title("Barplot")
plt.xlabel("Languages")
plt.ylabel("NO. Of students")
plt.xticks(n+width,classes)
```

```python
n=np.arange(len(classes))
width=0.2
plt.bar(n,classes_a,width=width)
plt.bar(n+width,classes_b,width=width)
plt.bar(n+width+width,classes_c,width=width)
plt.title("Barplot")
plt.xlabel("Languages")
plt.ylabel("NO. Of students")
plt.xticks(n+width,classes)
plt.show()
```

```python
n=np.arange(len(classes))
width=0.2
plt.bar(n,classes_a,width=width,bottom=classes_b+classes_c)
plt.bar(n,classes_b,width=width,bottom=classes_c)
plt.bar(n,classes_c,width=width)
plt.title("Barplot")
plt.xlabel("Languages")
plt.ylabel("NO. Of students")
plt.xticks(n,classes,rotation=20)
plt.show()
```
